```python
%matplotlib inline
import pandas as pd
import numpy as np
import scipy.stats as stats
from sklearn.cross_validation import cross_val_score, train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report
import matplotlib.pyplot as plt

# preparing data
from sklearn.metrics import precision_recall_fscore_support
from sklearn.preprocessing import LabelEncoder

gatrain=pd.read_csv("/home/danhd/PycharmProjects/TalkData/src/gender_age_train.csv")
gatest=pd.read_csv("/home/danhd/PycharmProjects/TalkData/src/gender_age_test.csv")
phone = pd.read_csv('/home/danhd/PycharmProjects/TalkData/src/phone_brand_device_model.csv',encoding='utf-8')

dup = phone.groupby('device_id').size()
dup = dup[dup > 1]
dup = phone.loc[phone.device_id.isin(dup.index)]
first = dup.groupby('device_id').first()
last = dup.groupby('device_id').last()
diff = (first != last).sum(axis=1).nonzero()
phone = phone.drop_duplicates('device_id', 'first')

# lebrand = LabelEncoder().fit(phone.phone_brand)
# phone['brand'] = lebrand.transform(phone.phone_brand)
# m = phone.phone_brand.str.cat(phone.device_model)
# lemodel = LabelEncoder().fit(m)
# phone['model'] = lemodel.transform(m)

#legender = LabelEncoder().fit(gatrain.gender)
#gatrain['gender'] = legender.transform(gatrain.gender)

# legroup = LabelEncoder().fit(gatrain.group)
# gatrain['group'] = legroup.transform(gatrain.group)

# train = gatrain.merge(phone[['device_id', 'brand', 'model']], how='left', on='device_id')

# c = train.groupby(['age','gender']).size().unstack().reindex(index=np.arange(train.age.min(), train.age.max()+1)).fillna(0)
# ax1, ax2 = c.plot(kind='bar',figsize=(12,6),subplots=True);


# predictors = ["age", "brand", "model"]

brand_dummies = pd.get_dummies(phone['phone_brand'])

brand_dummies['device_id'] = pd.Series(phone['device_id'], index=brand_dummies.index)

gatrain = gatrain.drop('age', 1).drop('group',1)
train = gatrain.merge(brand_dummies, how='left',on='device_id').drop('device_id',1)


alg = LogisticRegression(random_state=0)

scores = cross_val_score(alg, train.ix[:, train.columns != 'gender'], train["gender"], cv=3)

train_data, test_data = train_test_split(train, train_size=0.7)


# test_data = test_data.drop("gender", 1)

print "Logistic Regression Accuracy:", scores.mean()

# Train the algorithm using all the training data
def make_submission(alg):
    alg.fit(train_data.ix[:, train_data.columns != 'gender'], train_data["gender"])

    print "Model: {}".format(alg.coef_)    
    # Make predictions using the test set.
    predictions = alg.predict_proba(test_data.ix[:, test_data.columns != 'gender'])

    data = [x[1] for x in predictions]
    df=pd.DataFrame({'gender':test_data['gender'],'prob':data})
    male=df[df['gender']=='M'].prob
    female=df[df['gender']=='F'].prob
    
    dfMale = pd.DataFrame(male)
    dfMale.plot(kind='density')
    
    dfFemale = pd.DataFrame(female)
    dfFemale.plot(kind='density',color='red')

#     fig = plt.figure()
#     ax = fig.add_subplot(111)
#     numBins = 50
    
#     ax.hist(male,numBins,color='green',alpha=0.8,label='M')
#     ax.hist(female,numBins,color='red',alpha=0.8,label='F')
#     plt.show()
#     precision, recall, fscore, support = precision_recall_fscore_support(test_data["gender"], predictions)
#     print('precision: {}'.format(precision))
#     print('recall: {}'.format(recall))
#     print('fscore: {}'.format(fscore))
#     print('support: {}'.format(support))
    #print classification_report(test_data["gender"], predictions, digits=3)

make_submission(alg)
```
